```
#convention for importing numpy and pandas
import numpy as np
import pandas as pd
```

```
#dataset import
url='https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data
```

```
#Location of answer for Q4
fuel_data.describe(include='all') 
```

```
#group by the 'fuel_type_code_pudl' and find mean of 'fuel_cost_per_unit_burned'
fuel_data.groupby('fuel_type_code_pudl')[['fuel_cost_per_unit_burned']].mean()
```

```
#calc for skewness and kurtosis
a=fuel_data['fuel_qty_burned'].skew(axis=0)
b=fuel_data['fuel_qty_burned'].kurt(axis=0)
print('fuel qty burned skewness is ', a, 'fuel qty burned skewness is ', b)
```

```
#To find the sum of missing values
fuel_data.isnull().sum()
```

```
#missing values % calculation
(fuel_data['fuel_unit'].isnull().sum()/fuel_data[['fuel_unit']].shape[0])*100
```

```
#To find the correlation among the different columns
fuel_data.corr()
```

```
#To query the requisite colums and perform the mathematical operations
fuelcost98 = fuel_data.query('report_year == 1998 and fuel_type_code_pudl == "coal"')['fuel_cost_per_unit_burned'].sum()
fuelcost94 = fuel_data.query('report_year == 1994 and fuel_type_code_pudl == "coal"')['fuel_cost_per_unit_burned'].sum()
((fuelcost98-fuelcost94)/fuelcost94)*100
```

```
#To find mean of Fuel Cost per unit delivered
fuel_data.groupby("report_year")[["fuel_cost_per_unit_delivered"]].mean()
```
